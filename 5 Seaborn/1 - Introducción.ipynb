<code>

<ul>
<li><b>relplot (relacional):</b> scatterplot, lineplot.</li>
<li><b>displot (distribuciones):</b> histplot, kdeplot, ecdfplt, rugplot.</li> 
<li><b>catplot (categórica):</b> stripplot, swarmplot, boxplot, violinplot, pointplot, barplot.</li>  
</ul>
<h2>Barplot</h2>


```
# imports y datos
import seaborn as sns 
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.set(style="dark" , palette="dark" , font_scale=1)

sns.barplot( x=["A", "B" , "C"] , y=[1,3,2])

plt.show()
```

<h2>Stripplot y swarmplot

```
tips
```

```
# GRAFICO CATEGORICO STRIPPLOT SIN DODGE 

sns.stripplot( data=tips , x="day" , y="total_bill" , hue="sex" )
plt.show()
```

```
# GRAFICO CATEGORICO STRIPPLOT CON DODGE 'dodge=True'

sns.stripplot( data=tips , x="day" , y="total_bill" , hue="sex" , dodge=True)
plt.show()
# Dodge nos separa las variables de agrupamiento
```

```
# GRAFICO CATEGORICO 'SWARMPLOT'
sns.swarmplot( data=tips , x="day" , y="total_bill" , hue="sex" , dodge=False)
plt.show()
# Cuando utilizamos swarmplot los valores iguales quedan ALINEADOS
# En cambio cuando utilizamos stripplot los valores quedan AGRUPADOS.
```

<h2>Gráficas combinadas</h2>

```
# GRÁFICOS COMBINADOS

sns.boxplot( data = tips , x="day" , y="total_bill" , hue="sex", dodge=True )
sns.swarmplot( data=tips , x="day" , y="total_bill" , hue="sex" , dodge=True , color="white" ,marker="*")
plt.show()
# combina los dos graficos, es de caja y el swarplot
```

<h2>Violín</h2>
<p>Es una gráfica parecida a boxplot, pero en lugar de cuartiles, muestra la concentración de datos. Admite función <b>split</b> para comparar variables categóricas.</p>

```
# GRAFICO DE CATEGORIA VIOLINPLOT sin split
sns.violinplot(data=tips , x="day" , y="total_bill" , hue="sex")
plt.show()
```

```
# GRAFICO CATEGORICO VIOLINSPLOT CON SPLIT
sns.violinplot(data=tips , x="day" , y="total_bill" , hue="sex" , split=True)
plt.show()
```

<h2>Catplot</h2>
<p>Podemos combinar con otra variable usando <b>col</b> y seleccionar el tipo de gráfica con kind: </p>
<ul>
<li>strip (default)</li>
<li>swarm</li>
<li>box</li>
<li>violin</li>
<li>boxen</li>
<li>point</li>
<li>bar</li>
<li>count</li>
<ul>


```

sns.catplot( data=tips , x="day" , y="total_bill" , hue="sex" , dodge=True , kind="violin" , split=True , col="time")
plt.show()

# kind indica que tipo de gráfico quiero utilizar
# col me permite combinar otra variable al agrupamiento 
```

<h2>Pairplot</h2>
<p>Correalciones todas las variables numéricas entre sí.</p>
<p>Podemos eliminar las gráficas repetidas.</p>

```
sns.pairplot(tips)
plt.show()
# SOLO para variables numéricas
```

```
# CORNER: Eliminamos gráficas repetidas
sns.pairplot(tips , corner=True)
plt.show()
# hace que no devuelva los gráficos que son repetidos 'corner=True'
```

```
# imports y datos
import matplotlib.pyplot as plt 
import numpy as np


arts = ['PANTALÃ“N', 'REMERA', 'BLUSA', 'CAMISA', 'POLLERA', 'ABRIGO']
stock = [1524, 2684, 800, 1756, 1008, 3521]

plt.bar( arts , stock )
plt.show()
```

```
# COLORES Y TAMAÃ‘O
plt.bar( arts , stock , width=0.7 , color=["red" , "blue" , "green" , "yellow" , "violet" , "pink"] )
plt.show()
```

```
# cambiar nombre de las variables en el eje x con xticks y rotation
arts_2 = ['PANTALÃ“N', 'REMERA', 'BLUSA', 'CAMISA', 'POLLERA', 'ABRIGO' , 'PANTALÃ“N2', 'REMERA2', 'BLUSA2', 'CAMISA2', 'POLLERA2', 'ABRIGO2']
stock_2 = [1524, 2684, 800, 1756, 1008, 3521 ,1524, 2684, 800, 1756, 1008, 3521]

plt.bar( arts_2 , stock_2 )
plt.xticks(rotation=45)
plt.show()

# Rotation modifica la rotacion del texto de las grÃ¡ficas (rotacion de los labels)
```

```
# barh : Es el grÃ¡fico de barras horizontal 

plt.barh(arts , stock)
plt.show()
```

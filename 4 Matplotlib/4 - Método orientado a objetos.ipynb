<ul>
<li>Figure: es el lienzo donde voy a ubicar mis gráficas</li>
<li>Axes: son las gráficas individuales
</li>
</ul>

```python
# imports y datos
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(0,5,20)
y = x ** 2
```

```python
fig = plt.figure()  #creo el lienzo
axes = fig.add_axes([0.1,0.1,0.5,0.9])  #[posx,posy,sizex,sizey]
axes.plot(x,y,"b")
fig.show()
# axes es el grafico
```

```python
fig = plt.figure()#creo el lienzo
axes = fig.add_axes([0.1,0.1,0.7,0.9])#[posx,posy,sizex,sizey]
axes.plot(x,y,"b")


axes_dos = fig.add_axes([0.2,0.5,0.2,0.2])
axes_dos.plot(y,x,"r")
axes_dos.set_facecolor("yellow")
fig.show()
```

<p>Ambos axes son diferentes objetos dentro del mismo lienzo.</p>

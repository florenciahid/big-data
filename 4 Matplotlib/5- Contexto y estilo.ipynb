<h2>Contexto</h2>
<p>El contexto facilita la lectura de nuestras gráficas: etiquetas, leyendas y títulos</p>

```
# imports y datos
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0,5,20)
y = x ** 2
```

<h2>Estilos</h2>

```
print('¿Qué estilos hay disponibles en matplotlib.pyplot?')
print( plt.style.available ) 
# para saber todos los estilos disponibles para gráficos
# tienen una paleta de colores definida, son como un 'tema'
```

```
# UTILIZAR ESTILOS
plt.style.use("Solarize_Light2")
plt.plot(x,y)
plt.show()
```

```
# GROSOR DE LÍNEA (linewidth)
# TRANSPARENCIA (alpha)

plt.style.use("_mpl-gallery")
plt.plot(x,y, linewidth=1 )
plt.show()

```

```
# ESTILO DE LÍNEA : linestyle
# MARKERS : marker y markersize


plt.style.use("_mpl-gallery")
plt.plot(x,y, linestyle="dashed" , linewidth=1 ,marker="*" , markersize=5 )
plt.show()
```

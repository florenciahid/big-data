<h2>Subplot</h2>
<p>Permite crear gráficos dentro de una sola. Esto lo hace a través de una <b>matriz de gráficos</b> y se puede acceder a cada uno a través de sus índices. Es decir, me permite tener múltiples gráficos en uno solo.</p>


```
# imports y datos
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(0,5,11)   # np.linspace(min, max, cantidad de numeros)
y = x ** 2
```

```
# SINTAXIS: matplotlib.pyplot.subplot(filas,columnas,index)

plt.subplot(2,2,1) #Grafico 1
plt.plot(x,y , "r--")
plt.plot(y,x, "b:")


plt.subplot(2,2,2) #Grafico 2
plt.pie(y)

plt.subplot(2,2,3) #Grafico 3
plt.hist(y)


plt.subplot(2,2,4) #Grafico 4
plt.pie(x)

plt.show()

# Cuando trabajo con colores base, por ej "r"
# puedo hacer marker y lineas en la misma declaracion de color: "r--"
```

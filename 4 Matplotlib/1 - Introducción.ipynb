<h3>Pyplot</h3>
<p>Es la funcionalidad de Matplotlib para mostrar las gráficas</p>

```python
# Imports y datos
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 5, 10)
y = x**2

print(x)
print(y)
```

```python
plt.plot(x,y)
plt.show()

# plot me permite realizar graficos a partir de los valores (x, y)
```

<h4>Personalización: Colores</h4>
<small>Documentación: <a href='https://matplotlib.org/3.5.0/gallery/color/named_colors.html'>https://matplotlib.org/3.5.0/gallery/color/named_colors.html</a></small>

```python
plt.plot(x,y+2, "r")
plt.plot(x,y, "tab:purple")
plt.plot(x,y+5 , "darkblue")
plt.plot( x, y+7 ,"#4feb34")
plt.show()

# la tercer variable nos indica el color de cada gráfico.
# podemos observar diversas maneras de indicar color a un grafico: codigo, nombre, hexagesimal
```

<h4>Personalización: Markers</h4>
<small>Documentación: <a href='https://matplotlib.org/3.5.0/api/markers_api.html?highlight=markers#module-matplotlib.markers'>https://matplotlib.org/3.5.0/api/markers_api.html?highlight=markers#module-matplotlib.markers</a></small>
<p>Revisa la lista completa de markers en la documentación.</p>
<p>Algunos de los más utilizados:</p>
<img src ='img/markers.png'>


```python
# puedo incluir el marker con el color si uso base colors. Si además quiero la línea, pongo *-
plt.plot(x,y+2, "r*-")
plt.plot(x,y, "tab:purple" , marker="p")
plt.plot(x,y+5 , "darkblue" , marker="4")
plt.plot( x, y+7 ,color="#4feb34" , marker="s")
plt.show()

# los markers se indican en nuestro gráfico en los puntos correspondientes de los valores de nuestras variables
# los markers los podemos indicar en el tercer parametro (con el parametro color).
# tambien los podemos indicar como un nuevo parametro:  ' marker="4" '
```

<h4>Personalización: Estilos de líneas</h4>
<ul>
<li>Sólida: '-'</li>
<li>Guiones: '--'</li>
<li>Guión y punto: '-.'</li>
<li>Punteado: ':'</li>
<li>Ninguna: None, '', o ' '</li>
</ul>

```python
# puedo incluir el marker con el color si uso base colors. Si además quiero la línea, pongo *-

plt.plot(x,y+2, "y-")
plt.plot(x,y, "tab:purple" , linestyle=":")
plt.plot(x,y+5 , "darkblue" , linestyle="-.")
plt.plot( x, y+7 ,"#4feb34" , linestyle="--")
plt.show()


# la personalizacion de linea se indican a lo largo de toda la linea de nuestro gráfico 
# la personalizacion de linea la podemos indicar en el tercer parametro (con el parametro color).
# tambien los podemos indicar como un nuevo parametro:  ' linestyle=":" '
```

```python
# Puedo utilizar otros tipos de gráfica
plt.hist(y , color="#d69f51")
plt.show()

# en este caso utilizamos los valores de nuestra variable 'y'
```

```python
# ¿cómo buscar las opciones para cada gráfica en la documentación?
plt.scatter(x,y, marker="8" , color="#d69f51")
plt.show()

# nos indica los puntos de nuestras variable 'x' , 'y'
```

```python
# Distribución caja y bigotes
plt.boxplot(x)
plt.show()
```

```python
exp = (0,0,0.5,0,0,0,0,0,0,0)  # explode values
plt.pie(y , radius=1.5 , explode=exp)
plt.show()

# explode nos permite que uno de nuestros valores se ecuentre separado.
# en este caso el tercer valor está separado a 0.5 de los demás
```

```python
print("Labels y colores")
l = ("Anana" , "Naranja" , "Palta" , "Tomate")  # son los valores de los labels
sizes = [240,130,230,150]
e = (0,0,0.1,0)  # son los valores del explode
c = ["gold" , "orange" , "green" , "blue"]  # son los valores de los colores

plt.pie(sizes ,explode=e , labels=l ,colors=c)
plt.show()

# podemos customizar nuestro grafico opcionalmente
# el unico valor obligatorio es 'plt,pie' el esto son opcionales
```

```python
# VARIABLES AUTOPCT , STARTANGLE
# 'autopct' : Nos sirve para mostrar porcentajes 
# 'startangle' : Nos indica el angulo de inicio de nuestro gráfico

print("AUTOPCT: Mostrar porcentajes")  

l = ("Anana" , "Naranja" , "Palta" , "Tomate")
sizes = [240,130,230,150]
e = (0,0,0.1,0)
c = ["gold" , "orange" , "green" , "blue"]

plt.pie(sizes ,explode=e , labels=l ,colors=c , startangle=90 , autopct="%1.2f%%") 
plt.savefig("plot.png" , dpi=300)  # nos guarda la imagen del grafico
plt.show()

# %1.2f%% nos indica la cantidad de decimales, en este caso dos
# El valor que le indicamos a la variable 'startangle' es el valor del angulo del gráfico
# plt.savefig Guarda nuestro grafico, debemos escribirlo SIEMPRE antes de 'plt.show()'
# plt.savefig queda guardado en la carpeta 'files' de nuestro proyecto en deepnote
```

<h3>Guardando los gráficos</h3>
<p>Todos los gráficos que estuvimos haciendo se pueden guardar como PNG o JPG.</p>
<p>Esto lo hacemos con la siguiente función:</p>
<code>plt.savefig(nombre del archivo)</code>
<p>El archivo se agrega a la lista de Files del proyecto.</p>
<h4>Notas:</h4>
<p>Es muy imporante que se hagan antes de plt.show() ya que matplotlib habrá enviado el gráfico a la pantalla y este se limpia para hacer otro gráfico!</p>
<p>Se puede agregar el parámetro opcional dpi para obtener la resolución deseada (se especifica el número de puntos por pulgada):</p>
<code>plt.savefig('plot.png', dpi=300)</code>

<h3>Medidas de tendencia central en NumPy</h3>

```python
# MEDIA 

# Es el porcentaje de todos los valores de nuestro array o matriz

array_nros = np.array([5,5,5,7,8,10,12,12,14,15,16,16,16,20,2,2,2,3,4,22,22])
media = np.mean(array_nros)

print("Promedio/Media:" , f"{media:.2f}")

#cuando ingresamos " :.2f " nos indica que solo tomara DOS DECIMALES.
```

```python
# MEDIANA 

# En el valor que se encuantra en el centro en cuanto a posicion
# Si hay dos valores en el medio, se toma un porcertaje de los dos.

mediana = np.median(array_nros)
print("La mediana es: " , mediana)
```

```python
# CUARTILES

# son tres valores que nos dividen el dataset en cuatro partes
# 0.25 - 0.50 - 0.75  --> 0.50 es la mitad por eso igual a la mediana
# metodo quantile(arreglo , quantile)

q1 = np.quantile(array_nros , 0.25)
q2 = mediana
q3 = np.quantile(array_nros , 0.75)

print(f"Q1 = {q1} , Q2 = {q2} , Q3 = {q3}")
```

```python
# PERCENTIL 

#Es una medida de dispersión parecida a los cuartiles. Es más especifico
# divide la muestra (dataset) en 100 partes --> trabaja con porcentajes
# método percentile(arreglo, cuántos elementos tomar)

p60 = np.percentile(array_nros , 60)
p20 = np.percentile(array_nros , 20)
p50 = np.percentile( array_nros , 50)

print("Percentil 60:" , p60)
print("Percentil 20:" , p20)
print("Percentil 50:" , p50)
```

<h3>Universal functions (ufunc)</h3>
https://www.w3schools.com/python/numpy/numpy_ufunc.asp También ejecutan una operación elemento a elemento sobre un arreglo Reemplazan los operadores individuales. Son funciones nativas

Ejemplos:
<ul><li>np.add(vector,4) # suma 4</li>
<li>np.substract(vector,2) # resta 2</li>
<li>np.multiply</li>
<li>np.floor_divide(vector,3) # devuelve la division entera</li>
<li>np.negative(vector)</li>
<li>np.power(vector,2) #potencia</li>
<li>np.mod(vector,2)</li></ul>

```python
vector = np.array([(8, 9, 10), (11, 12, 13)])

print(vector.min())  # me devuelve el elemento MENOR de la matriz
print(vector.max())  # me devuelve el elemento MAYOR de la matriz
print(vector.sum())  # me devuelve la suma detodo los elementos de la matriz

print(np.sqrt(vector))  # me devuelve la raiz cuadrada de c/u de los valores del vector
print(np.negative(vector))  # me transforma todos los valores en negativo
# tanto 'np.sqrt(vector)' como 'n'p.negative(vector)' NO ALTERA la matriz original
# estan retornando una copia alterada
# si deseamos almacenarla debemos hacerlo en otra variable

print(np.power(vector, 2)) # me devuelve la potencia de c/u de los valores del vector
```

Puedo usar RANGOS de búsqueda, usando el mismo formato <code><mark>inicio:tope:paso</mark></code> de range.
Supongamos que quiero obtener los datos de las celdas:
<ul><li>naranja</li>
<li>turquesa</li>
<li>rojo</li>
<li>verde</li></ul>
<center><img src='img/slicing.png'></center>

```python
# CREO LA MATRIZ

matriz = np.array([[0,1,2,3,4,5],[10,11,12,13,14,15],[20,21,22,23,24,25],[30,31,32,33,34,35],[40,41,42,43,44,45],[50,51,52,53,54,55]])
print(matriz)

```

```python
# BUSCO EL RANGO EN NARANJA

# Fila 0 - Columnas 3 y 4 (la 5 no entra porque es el tope)
print(matriz[0, 3:5])

```

```python
# BUSCO EL RANGO TURQUESA

# Fila 4 y 5 - Columna 4 y 5
print(matriz[4: , 4:])
```

```python
# BUSCO EL RANGO ROJO

# Todas las filas - Columna 2
print(matriz[: , 2])
```

```python
# Si quiero que se imprima en columna, recuerdo que tengo el método reshape(filas,columnas)

# Todas las filas - Columna 2 (con reshape)
print(matriz[: , 2].reshape(6,1))
# utilizamos reshape para viasualizarlo como columna a los valores
```

```python
# BUSCO EL RANGO VERDE

# Filas 2 y 4 - Columnas 0, 2 y 4 (con step)
print(matriz[2::2 , ::2])

# matriz[inicio:fin:salto , :]  en este caso lo aplicamos en filas
#cuando indicamos ::2 hacemos referencia los saltos entre valores 
#en este caso son saltos de a dos 
```

<h3>Boolean Masking</h3>
Puedo obtener información sobre el contenido de mi arreglo, como valores repetidos, por encima o debajo de cierto valor, etc.

Podemos hacerlo mediante operadores mediante o ufuncs

```python
# CREO UNA MATRIZ CON VALORES ALEATORIOS, pero no quiero que cambien los valores cada vez que ejecuto mi código

np.random.seed(0)
arr_aleatorio = np.random.randint(0,100,(20,20))
print(arr_aleatorio)
 
# En este caso creamos un arreglo de 20x20 con num aleatorios entre 0 y 100
# np.random.seed nos permite que los num aleatorios sean siempre los mismos
# Cada vez que corremos nos dara siempre el mismo set aleatorio, NO SE ALTERA
```

```python
# Uso operadores para buscar

print("True donde el numero es < 50 ")
print( arr_aleatorio < 50)
```

```python
# numeros True dentro del arreglo
print(arr_aleatorio == 44)
print(arr_aleatorio == 67)
```

```python
# Con Ufuncs

# np.equal(array, valor)
# comprueba cada uno de los elementos si son iguales al valor - en este caso '30'

print(np.equal(arr_aleatorio, 30))
```

```python
#Con Ufuncs

# np.less(array, valor)
# comprueba si es mayor cada uno de los elementos a el valor - en este caso '32'
print(np.less(arr_aleatorio, 32))
```

```python
# Contando valores especificos en toda la matriz
# Cuenta todos los elementos que no son igual al valor - en este caso dif a '0' 
print(np.count_nonzero(arr_aleatorio))

# Contando valores específicos por fila, agregando la opción axis
print(np.sum(arr_aleatorio < 50 , axis=1)) 
# En este caso indica la suma de todos los valores MENORES a 50, en cada columna
# axis = 1 indica que realizamos la suma (de la condicion) en cada una de las columnas
```

<h3>Método copy</h3>
Permite crear copias de arrays.

```python
mi_array4 = np.array(([1,4,19,10,5],[3,9,10,8,8]))


# con el ID podemos observar que las dos variables referencian al mismo arreglo
# ambos apuntas a la misma direccion de memoria
mi_array5 = mi_array4

print(id(mi_array4))
print(id(mi_array5))


# En este caso podemos observarque el ID de ambas variables referencian diferentes arreglos
# ambos apuntan a diferentes direccion de memoria
mi_array6 = np.copy(mi_array4)

print(id(mi_array4))
print(id(mi_array6))
```

<h3>Arreglos de Booleanos</h3>
Con dtype puedo cambiar cómo se muestran los valores en pantalla

```python
booleanos1 = np.array([1,0,1,1,0,1,0,0])
booleanos2 = np.array([1,0,1,1,0,1,0,0], dtype=bool)

print( booleanos1 )
print( booleanos2 )
```

```python
booleanos1 = np.array([1,0,1,1,0,1,0,0])
booleanos3 = np.array([0,0,1,0,1,1,0,1])

booleanos4 = booleanos1 & booleanos3  # AND
print("And:" , booleanos4)

booleanos5 = booleanos1 | booleanos3 # OR
print("OR:" , booleanos5)


# Para poder aplicar '&' (and) , '|' (or) los arreglos debe ser del mismo tipo
# Se representan como booleanos, el 1 es Verdadero y el 0 es Falso
# Obtenemos es resultado de una CONJUNCION (and - &) o de una DISYUNCION (or - |)
```

```python
# Aplicando ufuncs obtengo los valores True y False

booleanos6 = np.logical_and(booleanos1 , booleanos3)
booleanos7 = np.logical_or(booleanos1 , booleanos3)

print("And:" , booleanos6 )

# otra manera de obtener valores booleanos con funciones nativas
```

<h3>all() y any()</h3>
Métodos para evaluar todos los valores de un arreglo

```python
# 'ALL' me permite evaluar si todos los valores son True
# 'ANY' me permite evaluar si hay algún valor True

booleanos1 = np.array([1,0,1,1,0,1,0,0])
booleanos8 = np.array([0,0,0,0,0,0,0,0])
booleanos9 = np.array([1,1,1,1,1,1,1,1])

print("Todos los valores del arreglo son True")
print(np.all(booleanos1))
print(np.all(booleanos8))
print(np.all(booleanos9))

print()

print("Algun valor del arreglo es True")
print(np.any(booleanos1))
print(np.any(booleanos8))
print(np.any(booleanos9))
```

<h3>Broadcasting</h3>
NumPy me permite operar entre arreglos de diferente estructura, transformándolos para que tengan el mismo tamaño para que no se generen errores.

Así es el proceso:
<center><img src='img/broadcasting.png'></center>

```python

vector_a = np.arange(0, 20, 5)  #np.arange(min, max, saltos)
print(vector_a)
```

```python
print(vector_a.shape)  #.shape nos permite ver la forma de nuestro arregloctor 
```

```python
# 'np.newaxis' es un metodo nos permite cambiar la orientacion de nuestro arreglo

vector_b = vector_a[:, np.newaxis]
print(vector_b)
print(vector_a.shape)

# nuestro arreglo paso de tener 1 fila y 4 columnas a tener 4 columnas y 1 fila

print(id(vector_a))
print(id(vector_b))
# tienen diferente direccion de memoria
```

```python
print( vector_a + vector_b )

# completo la fila, completo la columna y sumo a cada uno.
```

<h2>Medidas de tendencia central</h2>
<p>Son una manera de resumir información de toda una población y lo más básico en cuanto a análisis de datos.</p>


<h3>Moda</h3>
<p>Representa el dato que más se repite en una distribución de datos. Puede ser unimodal, bimodal o polimodal.</p>
<p><b>Ejemplo:</b> Buscamos qué cantidad de personas de cada edad hay en el aula.
<p>Si no hay gran cantidad de datos, pueden representarse en una <b>tabla de frecuencias.</b></p>
<p>Si el rango de datos es muy amplio, conviene utilizar un <b>diagrama de frecuencias.</b></p>
<center><img src="img/frecuencias.png"></center>
<p>La moda no aplica para datos numéricos continuos</p>

<h3>Media</h3>
<p>Es lo que conocemos comúnmente como <b>promedio</b></p>
<ul>
<li>Ventaja: Permite extraer información valiosa de los datos rápidamente.</li>
<li>Desventaja: Hay valores extremos que pueden distorsionar la información.</li>
</ul>
<p>Si tomamos como ejemplo el caso de las edades de las personas del aula, el promedio da 20,97. Este valor está más cercano a 21, que no representa a la mayoría de la muestra. 
<center><img src="img/coffeeshop.png">
<em><small>Foto de harith irfan wahid en Unsplash</small></em></center>
<p>Supongamos que nos piden calcular el promedio de salario de los 10 gerentes que están en este momento en la cafetería, y que por política de la empresa, todos los gerentes cobran $250.000. El cálculo del promedio sería sencillo: <b><code>10 x $250.000 / 10 = $250.000</code></b></p>
<p>¿Qué sucedería si de repente ingresa a la cafetería el accionista mayoritario de la empresa, que obtiene ganancias de la empresa por $3.500.000 al mes?</p>
<p>Ahora el cálculo sería: <b><code>[(10 x $250.000) + $3.500.000] / 11 = $545.454,54,</code></b> valor bastante lejano a la realidad. Por eso se dice que la media está <b>sesgada</b> por los valores atípicos.</p>


<h3>Mediana</h3>
Aquí se trata de obtener el valor que se encuentra exactamente en el medio del set de datos, <b>una vez ordenados.</b>
<ul>
<li>Cuando los datos son impares: es el dato que está justo en el centro.</li>
<li>Cuando los datos son pares: se hace un promedio de los dos datos que están en el centro.</li></ul>
<p>En el ejemplo de las edades, la mediana da 20:</p>
<center><img src="img/mediana.png"></center>
<p>En el ejemplo de los sueldos de los gerentes, la mediana da $250.000:</p>
<center><img src="img/mediana2.png"></center>
<p>Como podemos apreciar, la mediana nos da una idea más precisa, porque no se ve afectada por los valores de los extremos, y no interesa si se alejan demasiado de la mediana. Sería la opción más adecuada cuando nos encontramos con valores atípicos. Ésta es la diferencia fundamental entre la media y la mediana.</p>


### Ejemplos

```python
#IMPORT .CSV

import pandas as pd

df = pd.read_csv("data/cars_depurado.csv")
df

#Esto es un DATAFRAME:
#Es la estructura de datos principal de pandas
```

```python
# PROMEDIO PRECIOS
df["price_usd"].mean()
```

```python
# MEDIANA PRECIOS
df["price_usd"].median()
```

<p>Si quisiéramos ver la frecuencia de precios, nos encontraríamos con la dificultad de que los datos distintos son muy numerosos.</p>
<p>Para estos casos, se utiliza el <b>histograma</b>, una generalización del diagrama de frecuencias, que nos permite agrupar los datos en rangos.</p> 

```python
# MODA/FRECUENCIA PRECIOS
df["price_usd"].plot.hist()
```

```python
# SEABORN
import seaborn as sns 

sns.displot(df, x="price_usd" , hue="body_type" )

#"body_type" es el tipo de auto que es
#"hue" determina cuál es la columna que determina para contar
```

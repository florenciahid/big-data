<h2>Tipos de métricas descriptivas</h2>

<h3>1. Medidas de tendencia central</h3>
<p>Las medidas de tendencia central son medidas estadísticas que pretenden resumir en un solo valor a un conjunto de valores. Representan un centro en torno al cual se encuentra ubicado el conjunto de los datos. Las medidas de tendencia central más utilizadas son: <b>media, mediana y moda.</b> Nos indican cómo están distribuidos los datos.</p>
<center><img src="img/tend_central.png">
<em><small>Imagen: Allende H y AhumadaS, ILI-280</small></em></center>

<h3>2. Medidas de dispersión</h3>
<p>Las medidas de dispersión entregan información sobre la variación de la variable. Pretenden resumir en un solo valor la dispersión que tiene un conjunto de datos, teniendo en cuenta los valores obtenidos en el estudio de tendencia central. Las medidas de dispersión más utilizadas son: <b>rango de variación, varianza, desviación estándar y coeficiente de variación.</b> Indican cómo se alejan los datos respecto de la media aritmética y sirven como indicadores de la variabilidad de los datos.</p>

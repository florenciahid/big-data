<h2>Diccionarios</h2>

<h2 style='color:#FFFFFF'>JSON</h2>
<center><img src='img/json.png', width='300'></center>
<p>El formato JSON se utiliza para estructurar datos en forma de texto y permite el intercambio de información entre aplicaciones de manera sencilla, liviana y rápida.</p>
<center><img src='img/json2.png'></center>

```python
# JSON a dataframe
harry = pd.read_json("data/hpcharactersdataraw.json")
```

```python
# Cantidad de registros
print(harry["Name"].count() )
# Nombres de las columnas
print(harry.columns)
# Obtener todos datos de un registro
print( harry.iloc[5])
# Obtener un dato específico

print( harry.iloc[5]["Name"])
```

```python
# Un JSON también se puede levantar como un dataframe de series indexadas
harry_2 = pd.read_json("data/hpcharactersdataraw.json" , typ="Series")

harry_2
```

```python
# SLICING

harry[0:10]
```

```python
# Filtrado con Slicing

harry[0:5][ ["Name" , "Blood"] ]

# Para realizar el filtrado de dos columnas debemos realizarlo dentro de una lista
# por eso indicamos con doble corchete [["Name" , "Blood"]]
# Si filtramos solo una columna:   harry[0:5][ "Name" ]

# Esto sucede por como funcionan los ADS y los QUARTZ (empaquetador y desempaquetador)
```

<h3>Filtrado por loc/iloc</h3>

```python
# loc busca por 'nombre' del label y me toma el tope
harry.loc[0:5][ ["Name", "Blood"] ]
```

```python
# ¿Quiénes son Muggles?

muggles = harry[ harry["Blood"]  == "Muggle-born"]
muggles["Name"]

# Solo imprimimos los nombres de los que cumplen con la condicion "Muggle-born"
```

```python
muggles = muggles.drop_duplicates()
muggles
```

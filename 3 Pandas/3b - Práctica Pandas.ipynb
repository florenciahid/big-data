```
# imports y datos
import pandas as pd 

df = pd.read_csv("data/ventas.csv")

df
```

```
# EXPLORO LAS COLUMNAS DEL DF

print(df.columns)
```

```
# cuantos registros tiene el dataframe?

df["Importe"].count()
```

```
# qué tipo de datos tiene?
df.dtypes 

# nos indica qué tipo de dato son cada una de las culumnas de nuestro DataFrame
```

```
#CONVIERTO LA COLUMNAS DE importes a FLOAT

df["Importe"] = df["Importe"].astype(float)

df["Comision_x_venta"] =df["Comision_x_venta"].astype(float)
```

```
# VERIFICO
# qué tipo de datos tiene ahora?
df.dtypes
```

```
# LE DOY FORMATO DE $$$$
pd.options.display.float_format = '$ {:,.2f}'.format

df["Importe"]

# .2f nos indica que es un tipo de dato FLOTANTE con DOS DECIMALES.
```

```
#ORDENAR EL DATAFRAME POR NOMBRE DEL VENDEDOR

df.sort_values("Vendedor")
```

```
df
```

```
#OBTENER EL MÁXIMO DE TODAS LAS COLUMNAS

df.max()
```

```
# ACTIVIDAD DEL VENDEDOR Mufi Panchin con boolean masking

df["Vendedor"] == "Mufi Panchin"  #nos devuelve resultados booleanos

df[ df["Vendedor"] == "Mufi Panchin"] 
# nos devuelve los resultatos que coinciden con el termino indicado.

```

```
# obtengo los registros completos de actividad de Mufi Panchin

df_2 = df.loc[ df["Vendedor"]  == "Mufi Panchin"]

df_2
```

```
# BUSCO LOS REGISTROS DONDE EL VENDEDOR SEA "Mufi Panchin" y Sede sea "Paternal"

resultado = df[ (df["Vendedor"] == "Mufi Panchin") & ( df["Sede"] == "Paternal") ]
resultado

# si quiesieramos saber el máximo:
# resultado = df[ (df["Vendedor"] == "Mufi Panchin") & ( df["Sede"] == "Paternal") ].max()
```

```
# CARGO LAS MEJORES VENTAS (>200000) EN OTRO DATAFRAME
mejores = df[ df["Importe"] > 200000 ]

# Y BUSCO A MUFI PANCHIN en el nuevo df
df_3 = mejores[ mejores["Vendedor"] == "Mufi Panchin"]

df_3
```

```
# cuántas operaciones >200000 concretó Mufi Panchin?
print("Total de operaciones: " , df_3["Importe"].count() )
```

```
# QUIERO VER QUÉ RUBROS VENDIÓ MUFI PANCHIN Y POR QUÉ IMPORTE

print( df_3["Rubro"] , df_3["Importe"])
```

```
print( df_3[ ["Vendedor" , "Importe" ]].sort_values("Importe") )
```

```
# si el importe es menor a 350000, cambiarlo a None(ninguno)

df_3.loc[ df_3['Importe'] < 350000 ] = None
print( df_3[ ["Rubro" , "Importe"]])
```

```
# cuando haga un count, Los None van a ser ignorados
print("Total de operaciones de Mufi Panchin por mas de 350.000: " , df_3["Importe"].count())
```
